```python
#tut-03.5

n=int(r.randint(1,10))
p=int(r.randint(1,10))
m=int(r.randint(1,10))

D=np.random.random((n,p))
E=np.random.random((p,m))
F=np.dot(D,E)

print(D,"\n","\n","\n",E,"\n","\n","\n",)
print("Dot Product: ","\n","\n","\n", F)
```
